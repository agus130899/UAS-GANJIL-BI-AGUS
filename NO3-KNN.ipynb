```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import math 
import pickle
```

```python
df = pd.read_excel('caesarian.xlsx')
df.head()
```

```python
dis = []
for i in range(80):
    dis.append(math.sqrt((float(df.iloc[i]['Usia'])-25)**2+
                         (float(df.iloc[i]['Kelahiran_ke-'])-1)**2+
                         (float(df.iloc[i]['Waktu_Kelahiran'])-0)**2+
                         (float(df.iloc[i]['Tekanan_darah'])-0)**2))
```
